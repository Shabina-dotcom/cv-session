```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('./media/einstein.jpg', 0)

gblur5 = cv.GaussianBlur(img, (5, 5), 0)
gblur9 = cv.GaussianBlur(img, (9, 9), 0)


edge_img = cv.subtract(gblur5, gblur9)

plt.figure()
plt.imshow(edge_img), plt.title("edge detection"), plt.axis('off')
plt.show()
```
