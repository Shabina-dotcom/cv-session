```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1

```python
# Reading the image
img = cv2.imread('einstein.jpg')
```

```python
blurred1 = cv2.GaussianBlur(img,(5,5),2)
blurred2 = cv2.GaussianBlur(img,(9,9),3)
```

```python
plt.figure(dpi=200),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)
plt.figure(dpi=200),plt.axis("off"),plt.title("Gaussian Blur (5x5)"),plt.imshow(blurred1)
plt.figure(dpi=200),plt.axis("off"),plt.title("Gaussian Blur (9x9)"),plt.imshow(blurred2);
```

```python
edge_image = blurred1 - blurred2  #cv2.subtract(blurred1,blurred2)
```

```python
plt.figure(dpi=200),plt.axis("off"),plt.title("Edge Detection"),plt.imshow(edge_image);
```

```python
vid = cv2.VideoCapture('barrel.mp4')
```

```python
ret,frame = vid.read()
```

```python
hsv_target = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
```

```python
obj = cv2.imread('object image.png')
hsv_obj = cv2.cvtColor(obj,cv2.COLOR_BGR2HSV)
```

```python
plt.figure(dpi=200),plt.axis("off"),plt.title("HSV Object Image"),plt.imshow(hsv_obj);
plt.figure(dpi=200),plt.axis("off"),plt.title("HSV Target Image"),plt.imshow(hsv_target);
```

```python
hist_obj = cv2.calcHist([hsv_obj],[0], None, [256], [0,256])
```

```python
plt.figure()
plt.plot(hist_obj),plt.xlabel('Pixel intensity value'),plt.ylabel('Frequency'),plt.title('Histogram of Object Image')
plt.show()
```

```python
res = cv2.calcBackProject([hsv_target],[0],norm_hist,[0,256],1)
```

```python
disc = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
img = cv2.filter2D(res, -1, disc, res) #res is the matrix obtained after back projection
```

```python
thresh = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,0)
```

```python
final = cv2.merge((thresh,thresh,thresh))
```

```python
result = cv2.bitwise_or(hsv_target, final)
```

```python
plt.figure(dpi=100),plt.axis("off"),plt.title("Result"),plt.imshow(result);
```
