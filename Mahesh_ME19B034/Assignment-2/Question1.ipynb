```python
## Task 1: Edge Detection using Difference of Gaussians

import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('einstein.jpg', 1)
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred1 = cv2.GaussianBlur(imggray, (5, 5), 0)
blurred2 = cv2.GaussianBlur(imggray, (9, 9), 0)
blurred = blurred2-blurred1

plt.figure(),plt.imshow(img,cmap='gray'),plt.axis('off'),plt.title('Original')
plt.figure(),plt.imshow(blurred1,cmap='gray'),plt.axis('off'),plt.title('GaussianBlurred(5x5 Kernel)')
plt.figure(),plt.imshow(blurred2,cmap='gray'),plt.axis('off'),plt.title('GaussianBlurred(9x9 Kernel)')
plt.figure(),plt.imshow(blurred,cmap='gray'),plt.axis('off'),plt.title('Edge Detected image')
plt.show()
```
