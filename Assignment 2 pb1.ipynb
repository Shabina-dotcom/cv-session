```python
import cv2
from matplotlib import pyplot as plt
img=cv2.imread(r"C:\Users\abhil\Downloads\einstein.jpg")
img1 = cv2.GaussianBlur(img, (5,5), 0)
img2 = cv2.GaussianBlur(img, (9,9), 0)
img_final = img1-img2
plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)
plt.figure(),plt.axis("off"),plt.title("Edge detection using difference of Gaussians"),plt.imshow(img_final)
plt.show()
```
