```python
#Edge detection using Gaussian filters
#importing required libraries
import cv2
from matplotlib import pyplot as plt
#reading image
img=cv2.imread('media/einstein.jpg',0)
#creating edge detector
edg=cv2.GaussianBlur(img,(5,5),1)-cv2.GaussianBlur(img,(9,9),0)
#plotting images
plt.figure()
plt.imshow(img,cmap='gray'),plt.axis("Off"),plt.title("Original image")
plt.figure()
plt.imshow(edg,cmap='gray'),plt.axis("Off"),plt.title("Edges")
plt.show()
```
