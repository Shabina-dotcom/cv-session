```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('einstein.jpg',0)
#Reading the image

blur_gaus1 = cv2.GaussianBlur(img,(5,5),0.99)
#Plotting the images
blur_gaus2 = cv2.GaussianBlur(img,(9,9),2,3)

plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img,cmap='gray')
plt.figure(),plt.axis("off"),plt.title("Blurred Image1"),plt.imshow(blur_gaus1,cmap='gray')
plt.figure(),plt.axis("off"),plt.title("Blurred Image2"),plt.imshow(blur_gaus1,cmap='gray')
plt.show()
```

```python
#edge detection
edge = cv2.absdiff(blur_gaus1,blur_gaus2)
plt.axis('off'),plt.imshow(edge)
```
