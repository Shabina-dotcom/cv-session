```python
import cv2
#import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('media/einstein.jpg')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(img.shape)
```

```python
blur_gaus1 = cv2.GaussianBlur(img,(5,5),0)
```

```python
blur_gaus2 = cv2.GaussianBlur(img,(9,9),0)
```

```python
DOG = blur_gaus2 - blur_gaus1
```

```python
plt.figure(),plt.axis("off"),plt.title("Edge Detection using DoG"),plt.imshow(DOG)
```
