```python
import cv2
```

```python
img = cv2.imread('einstein.jpg')
```

```python
import matplotlib.pyplot as plt
# load image using cv2....and do processing.
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# as opencv loads in BGR format by default, we want to show it in RGB.
plt.show()
```

```python
import numpy as np
```

```python
blur_gaus = cv2.GaussianBlur(img,(11,11),0)
#Plotting the images

plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)
plt.figure(),plt.axis("off"),plt.title("Blurred Image"),plt.imshow(blur_gaus)
plt.show()
```

Clearly we see image gets blurred from the edges

```python
blur_gaus5x5 = cv2.GaussianBlur(img,(5,5),0)
blur_gaus9x9 = cv2.GaussianBlur(img,(9,9),0)
plt.figure(),plt.axis("off"),plt.title("Blurred Image using 5x5"),plt.imshow(blur_gaus5x5)
plt.figure(),plt.axis("off"),plt.title("Blurred Image using 9x9"),plt.imshow(blur_gaus9x9)
plt.show()
```

```python
edge_detection_image= blur_gaus5x5 - blur_gaus9x9
plt.figure(),plt.axis("off"),plt.title("Edge Detected image"),plt.imshow(edge_detection_image)
plt.show()
```

Hence this is the edge detected image

### Using two different standard deviations of 5x5 matrix

```python
blur_gaus5x5_sd0 = cv2.GaussianBlur(img,(5,5),0)
blur_gaus5x5_sd1 = cv2.GaussianBlur(img,(5,5),1)
plt.figure(),plt.axis("off"),plt.title("Blurred Image using 5x5 std=0"),plt.imshow(blur_gaus5x5_sd0)
plt.figure(),plt.axis("off"),plt.title("Blurred Image using 5x5 std=1"),plt.imshow(blur_gaus5x5_sd1)
plt.show()
```

finding the difference

```python
edg_img = blur_gaus5x5_sd0 - blur_gaus5x5_sd1
```

```python
plt.figure(),plt.axis("off"),plt.title("Edge img"),plt.imshow(edg_img)
plt.show()
```

Almost all edges are detected and the noise is reduced to almost 0

##### Now changing the std

```python
blur_gaus5x5_sd2 = cv2.GaussianBlur(img,(5,5),2)
plt.figure(),plt.axis("off"),plt.title("Blurred Image using 5x5 std=0"),plt.imshow(blur_gaus5x5_sd0)
plt.figure(),plt.axis("off"),plt.title("Blurred Image using 5x5 std=2"),plt.imshow(blur_gaus5x5_sd2)
plt.show()
```

```python
edg_img_2 = blur_gaus5x5_sd0 - blur_gaus5x5_sd2
plt.figure(),plt.axis("off"),plt.title("Edge img 2"),plt.imshow(edg_img_2)
plt.show()
```

Comparing all 3 images together

```python
plt.figure(),plt.axis("off"),plt.title("Edge img using two different kernels"),plt.imshow(edge_detection_image)
plt.figure(),plt.axis("off"),plt.title("Edge img using two different std(0,1)"),plt.imshow(edg_img)
plt.figure(),plt.axis("off"),plt.title("Edge img using two different std(0,2)"),plt.imshow(edg_img_2)
plt.show()
```
